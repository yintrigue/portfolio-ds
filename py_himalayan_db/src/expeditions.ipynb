```python
# load raw data frames
peaks_df = loaders.load_peaks(DATA_PATH)
exp_df = loaders.load_expeditions(DATA_PATH)
mem_df = loaders.load_members(DATA_PATH)
mem_unique_df = loaders.load_unique_members(mem_df=mem_df)
mem_all_df = loaders.load_members(DATA_PATH, False)

peaks_low_df = loaders.load_peaks(DATA_PATH, False)
exp_low_df = loaders.load_expeditions(DATA_PATH, False)
mem_low_df = loaders.load_members(DATA_PATH, False)
mem_unique_low_df = loaders.load_unique_members(mem_df=mem_df, eight_thousanders=False)

# plot color constants
# https://seaborn.pydata.org/tutorial/color_palettes.html
SNS_COLOR_PALETTE = "hls"
peaks_count_df = exp_df.loc[:,"peakid"].reset_index().groupby("peakid").count()
PEAK_COLOR_SNS = dict(zip(peaks_count_df.index, 
                          sns.color_palette(SNS_COLOR_PALETTE, 
                                            peaks_count_df.shape[0])))

# - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - 
# Add New Columns

# Reason Terminated, [termreason]:
#     0 – Unknown
#     1 – Success (main peak)
#     2 – Success (subpeak)
#     3 – Success (claimed)
#     4 – Bad weather (storms, high winds)
#     5 – Bad conditions (deep snow, avalanching, falling ice, or rock)
#     6 – Accident (death or serious injury)
#     7 – Illness, AMS, exhaustion, or frostbite
#     8 – Lack (or loss) of supplies or equipment
#     9 – Lack of time
#     10 – Route technically too difficult, lack of experience, strength, or motivation
#     11 – Did not reach base camp
#     12 – Did not attempt climb
#     13 – Attempt rumored
#     14 – Other

# create masks for exp results
def add_exp_results(exp_df: pd.DataFrame) -> pd.DataFrame:
        
    def r(codes: List[int], df) -> List[bool]:
        """Mask for [termreason].
        """
        r = df.loc[:,"termreason"]
        for i, c in enumerate(codes):
            if i == 0:
                r = (r == c)
            else:
                r = r | (r == c)
        return r

    mask_s_peak = r([1, 3], exp_df)
    mask_f_nature = r([4, 5], exp_df)
    mask_f_human = r([8, 9, 10], exp_df)
    mask_f_accident =  r([6], exp_df)
    mask_f_ill = r([7], exp_df)
    mask_others = r([0, 2, 11, 12, 13, 14], exp_df)

    # add [expr] & [expr_d]
    # [expr]
    #     s - success
    #     f - fail
    #     o - others (e.g. no summit attempt, unknown result, etc.)
    # [expr_d]
    #     s - success
    #     fn - fail due to unctrollable factors by nature
    #     fh - fail due to controllable human factors
    #     fa - fail due to accident including death
    #     fi - fail due to illness
    #     o - no success due to other reasons
    exp_df["expr"] = "s"
    exp_df.loc[mask_f_nature | mask_f_human | 
               mask_f_accident | mask_f_ill, "expr"] = "f"
    exp_df.loc[mask_others, "expr"] = "o"

    exp_df["expr_d"] = "s"
    exp_df.loc[mask_f_nature, "expr_d"] = "fn"
    exp_df.loc[mask_f_human, "expr_d"] = "fh"
    exp_df.loc[mask_f_accident, "expr_d"] = "fa"
    exp_df.loc[mask_f_ill, "expr_d"] = "fi"
    exp_df.loc[mask_others, "expr_d"] = "o"
    exp_df.loc[mask_f_nature, "expr_d"] = "fn"
    
    return exp_df

exp_df = add_exp_results(exp_df)
exp_low_df = add_exp_results(exp_low_df)

# - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - 
# summary
utils.pstr(exp_df)
```

```python
# [expr]
#     s - success
#     f - fail
#     o - others (e.g. no summit attempt, unknown result, etc.)

def waffle_results(exp_df: pd.DataFrame, title: str) -> None:
    # total failure rate
    df = exp_df.loc[:, ["peakid", "expid", "expr"]]
    df = df.groupby(["expr"]).count()["peakid"]
    df = df / sum(df) * 100
    df = round(df / sum(df) * 100).astype("int")

    data = {"Success": df.loc["s",], 
            "Failure": df.loc["f",], 
            "Unknown": df.loc["o",]}
    
    data["Success"] -= 1
    data["Unknown"] += 1
    
    fig = plt.figure(
        FigureClass=Waffle, 
        rows=5, 
        values=data, 
        colors=("#b0cb66", "#ca6d67", "#bbbbbb"),
        legend={"loc": "upper left", "bbox_to_anchor": (1, 1), 
                "framealpha": 0.25, "handletextpad": 1},
        labels=["{0} ({1}%)".format(k, v) for k, v in data.items()],
        icons=["hiking","user-injured","question-circle"], 
        icon_size=24,
        font_size=14, 
        figsize=(10, 3),
        icon_legend=True
    )
    fig.suptitle(title)

waffle_results(exp_df, "Expedition Failure Rate\n(Eight-Thousanders, 1905-2019)")
waffle_results(exp_low_df, "Expedition Failure Rate\n(Peaks Below 8,000m, 1905-2019)")
```

```python
def waffle_deaths(exp_df: pd.DataFrame) -> None:
    return sum(exp_df.loc[:, "mdeaths"]) / sum(exp_df.loc[:, "totmembers"])
    
# death ratio, eight-thousanders:non-eight-thousanders
r = waffle_deaths(exp_df) / waffle_deaths(exp_low_df)
print("The climber death rate on eight-thousanders is {:.2f}% more than peaks below 8,000 meters,".format((r - 1) * 100))
```

```python
deaths_df = mem_df.loc[mem_df.death, ["peakid", "calcage", "deathhgtm", "deathtype"]]
deaths_df = deaths_df.groupby("deathtype").count()
deaths_df = deaths_df.sort_values(by="peakid", ascending=False)["peakid"]

def death_decode(code:int) -> str:
    """Decode peaks.pstatus.
    """
    try:
        switch = {0: "Unspecified",
                  1: "Altitude Sickness",
                  2: "Exhaustion",
                  3: "Exposure",
                  4: "Fall",
                  5: "Crevasse",
                  6: "Icefall",
                  7: "Avalanche",
                  8: "Falling Rock/Ice",
                  9: "Disappearance",
                  10: "Illness",
                  11: "Other",
                  12: "Unknown"
                 }
        return switch[code]
    except Exception:
        return ""

deaths_df = deaths_df.reset_index()
deaths_df["deathtype"] = deaths_df["deathtype"].apply(death_decode)
deaths_df = deaths_df.set_index("deathtype")

# bar chart
fig, ax = plt.subplots(figsize=(FIG.W/1.5, FIG.H))
fig.suptitle("Death Types (1905-2019)", 
             fontsize=FIG.TITLE_FONT_SIZE)
fig.subplots_adjust(top=FIG.TITLE_LINESPACE)

sns.barplot(x=deaths_df.index, y="peakid", data=deaths_df, ax=ax,
            palette=sns.color_palette("Set2"))

ax.set_xlabel("Death Type", fontsize=FIG.FONT_SIZE)
ax.set_ylabel("Death Count", fontsize=FIG.FONT_SIZE)
ax.set_xticklabels(ax.get_xticklabels(), rotation = 90)

# patch with bar width
utils.sns_patch_bartchart_set_width(ax, .1)

pass
```
